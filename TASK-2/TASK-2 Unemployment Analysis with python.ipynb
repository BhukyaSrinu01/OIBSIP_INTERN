```python
# Changing the Attribute Names
data.rename(columns={"Region":"States"," Date":"Date"," Estimated Unemployment Rate (%)":"Estimated Unemployment Rate (%)",
                  " Estimated Employed":"Estimated Employed"," Estimated Labour Participation Rate (%)":
                   "Estimated Labour Participation Rate (%)"},inplace=True)
```

```python
data.head()
```

```python
# Changing the Data-Type of Date Attribute.
data['Date'] = pd.to_datetime(data['Date'],dayfirst=True)
```

```python
# Extracting Month from the Date Attribute
data["Month"] = data["Date"].dt.month
```

```python
# Updating Months to Months Name
data['Month_Name'] =  data['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
# Dropping the Months Attribute
data.drop(columns='Month',inplace=True)
```

```python
# Checking Unique Months
data['Month_Name'].unique()
```

```python
# Extracting Years From the Date Attribute
data['Year'] = data['Date'].dt.year
```

# Exploratory Data Analysis (EDA)

Visualizing Average Employement Rate State Wise.

```python
x = data.groupby(["States"])["Estimated Unemployment Rate (%)"].mean().sort_values(ascending=False).to_frame()

plt.figure(figsize=(15, 6))
sns.barplot(x=x.index, y="Estimated Unemployment Rate (%)", data=x, palette=sns.color_palette("Reds", 28)[::-1])
plt.title("State wise Average Unemployment Rate", fontweight="black", fontsize=20, pad=10)
plt.xticks(rotation=90)

plt.show()
```

Visualizing State Wise Average Labour Particiaption Rate.

```python
x = data.groupby(["States"])["Estimated Labour Participation Rate (%)"].mean().sort_values(ascending=False).to_frame()

plt.figure(figsize=(15, 6))
sns.barplot(x=x.index, y="Estimated Labour Participation Rate (%)", data=x, palette=sns.color_palette("Reds", 28)[::-1])
plt.title("State wise Average Labour Participation Rate", fontweight="black", fontsize=20, pad=10)
plt.xticks(rotation=90)

plt.show()
```

Visualizing State Wise Total Estimated Employed

```python
x = data.groupby(["States"])["Estimated Employed"].sum().sort_values(ascending=False).to_frame()

plt.figure(figsize=(15, 6))
sns.barplot(x=x.index, y="Estimated Employed", data=x, palette=sns.color_palette("Reds", 28)[::-1])
plt.title("State wise Total Estimated Employed", fontweight="black", fontsize=20, pad=10)
plt.xticks(rotation=90)

plt.show()
```

Visualizing Average "Unemployment Rate" & "Labour Participation" Rate w.r.t Area.

```python
x = data.groupby(["States"])["Estimated Unemployment Rate (%)"].mean().sort_values(ascending=False).to_frame()
z = data.groupby(["Area"])["Estimated Labour Participation Rate (%)"].mean().sort_values(ascending=False).to_frame()

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.barplot(x=x.index, y="Estimated Unemployment Rate (%)", data=x, palette=sns.color_palette("Reds", 3)[::-1])
plt.title("State wise Average Unemployment Rate", fontweight="black", fontsize=15, pad=10)
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
sns.barplot(x=z.index, y="Estimated Labour Participation Rate (%)", data=z, palette=sns.color_palette("Reds", 3)[::-1])
plt.title("State wise Average Labour Participation Rate", fontweight="black", fontsize=15, pad=10)
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

Visualizing Unemployment and Labour Rate throughout the years

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
ax = sns.lineplot(x="Year", y="Estimated Unemployment Rate (%)", data=data, color="red")
ax.set_xticklabels(["", "2019", "", "", "", "", "2020"])
plt.title("Unemployment Rate Over the Years", fontweight="black", fontsize=20, pad=10)

plt.subplot(1, 2, 2)
bx = sns.lineplot(x="Year", y="Estimated Labour Participation Rate (%)", data=data, color="blue")
bx.set_xticklabels(["", "2019", "", "", "", "", "2020"])
plt.title("Labour Participation Rate Over the Years", fontweight="black", fontsize=20, pad=10)

plt.tight_layout()
plt.show()
```

Visualizing Unemployment & Labour Rate in 2019 throughout the months

```python

plt.subplot(1, 2, 1)
sns.lineplot(x="Month_Name", y="Estimated Unemployment Rate (%)", data=data, color="red")
plt.title("Unemployment Rate over the Months in 2019", fontweight="black", fontsize=15, pad=10)
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.lineplot(x="Month_Name", y="Estimated Labour Participation Rate (%)", data=data, color="blue")
plt.title("Labour Participation Rate over the Months in 2019", fontweight="black", fontsize=15, pad=10)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
